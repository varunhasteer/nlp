```python
import numpy as np
import pandas as pd
import xml.etree.ElementTree as ET
from bs4 import BeautifulSoup
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline
tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")
nlp = pipeline("ner", model=model, tokenizer=tokenizer)
```

```python
conservative_d = ET.parse('/PE-Conservatives-Donation-Subscription.xml')
conservative_nd = ET.parse('/PE-Conservatives-NoDonation-Subscription.xml')
greenparty_d = ET.parse('/PE-GreenParty-Donation-Subscription.xml')
greenparty_nd = ET.parse('/PE-GreenParty-NoDonation-Subscription.xml')
liberal_d = ET.parse('/PE-Liberal-Donation-Subscription.xml')
liberal_nd = ET.parse('/PE-Liberal-NoDonation-Subscription.xml')
ndp_d = ET.parse('/PE-NDP-Donation-Subscription.xml')
ndp_nd = ET.parse('/PE-NDP-NoDonation-Subscription.xml')

conservative_d_root = conservative_d.getroot()
conservative_nd_root = conservative_nd.getroot()
greenparty_d_root = greenparty_d.getroot()
greenparty_nd_root = greenparty_nd.getroot()
liberal_d_root = liberal_d.getroot()
liberal_nd_root = liberal_nd.getroot()
ndp_d_root = ndp_d.getroot()
ndp_nd_root = ndp_nd.getroot()
```

```python
result_conservative_d = []
for x in conservative_d_root.iter('text'):
    result_conservative_d.append(nlp(x.text))
```

```python
result_conservative_nd = []
for x in conservative_nd_root.iter('text'):
    result_conservative_nd.append(nlp(x.text))
```

```python
result_greenparty_d = []
for x in greenparty_d_root.iter('text'):
    result_greenparty_d.append(nlp(x.text))
```

```python
result_liberal_d = []
for x in liberal_d_root.iter('text'):
    result_liberal_d.append(nlp(x.text))
```

```python
flat_list_result_conservative_d = [item for sublist in result_conservative_d for item in sublist]

list_names_conservative_d = []

length_conservative_d = len(flat_list_result_conservative_d)

for element in range(0,length_conservative_d):
  if(flat_list_result_conservative_d[element]['entity'] == 'B-PER'):
    list_names_conservative_d.append(flat_list_result_conservative_d[element]['word'])

newlist_alpha = [s for s in list_names_conservative_d if s.isalpha()]

names_conservative_d = set(newlist_alpha)
print("The names in Conservatives donation xml are : ",names_conservative_d)
```
